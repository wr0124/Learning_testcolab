```python
import torch

def add_constant_to_image(image, mask, constant_value=10):
    """Simple example function: adds a constant value to the image where the mask is non-zero."""
    return image + (mask * constant_value)

def apply_add_constant_to_5d(gt_image, mask, constant_value):
    """Apply the `add_constant_to_image` function to each image in a 5D tensor."""
    b, f, c, h, w = gt_image.shape  # Assuming 5D tensor: (batch, frames, channels, height, width)

    # Initialize an empty list to store results
    result = []

    # Iterate over the batch and frame dimensions
    for batch_idx in range(b):
        frames = []
        for frame_idx in range(f):
            # Extract 4D slices from the 5D tensors (single image)
            gt_image_4d = gt_image[batch_idx, frame_idx, :, :, :].unsqueeze(0)  # shape: (1, c, h, w)
            mask_4d = mask[batch_idx, frame_idx, :, :, :].unsqueeze(0)  # shape: (1, c, h, w)
            print( "gt ", gt_image_4d.shape)
            print( "mask ", mask_4d.shape)
            # Apply the example operation (add a constant value)
            cond_image_4d = add_constant_to_image(gt_image_4d, mask_4d, constant_value)

            # Append the result for this frame
            frames.append(cond_image_4d)
            print( "frames ", len(frames) )
        # Stack the frames back into a 5D structure
        result.append(torch.stack(frames, dim=1))  # Rebuild the frame dimension (f)
        print( "result ", len(result) )
    # Stack the batch back into a 5D tensor
    return torch.cat(result, dim=0)  # shape: (b, f, c, h, w)

# Example usage:

# Create random 5D tensors for gt_image and mask (batch, frames, channels, height, width)
gt_image = torch.randn(2, 3, 1, 2, 2)  # shape: (2, 3, 1, 5, 5) - batch=2, frames=3, channels=1, height=5, width=5
mask = torch.ones(2, 3, 1, 2, 2)  # shape: (2, 3, 1, 5, 5) - same shape as gt_image

# Apply the operation to each 4D slice
constant_value = 100
cond_image_5d = apply_add_constant_to_5d(gt_image, mask, constant_value)

# Print the resulting 5D tensor
print("Original 5D Tensor (gt_image):", gt_image)
print("Mask 5D Tensor:", mask)
print("Conditioned 5D Tensor (cond_image):", cond_image_5d)
```
